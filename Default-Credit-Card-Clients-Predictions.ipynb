```python
X = dataset.drop(['ID','default payment next month'],axis=1)
X
```

```python
y = dataset['default payment next month']
y
```

```python
dataset = pd.get_dummies(dataset,columns=['EDUCATION','MARRIAGE','SEX'],prefix=['EDU',"MRG",('MALE','FEMALE')])
```

```python
dataset
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=0)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print("LOGISTIC REGRESSION ALGORITHM")
print()
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()
lg.fit(X_train,y_train)
y_pred = lg.predict(X_test)
print('y_pred')
print(y_pred)
print()
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm = confusion_matrix(y_test,y_pred)
print('conufusion_matrix')
print()
print(cm)
print()
ac = accuracy_score(y_test,y_pred)
print('accuracy_score')
print(ac)
print()
cr = classification_report(y_test,y_pred)
print('classification_report')
print(cr)
print()
bias = lg.score(X_train,y_train)
print('bias=',bias)
print()
variance = lg.score(X_test,y_test)
print('variance=',variance)
```
